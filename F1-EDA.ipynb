```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'f1-data/winners.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')
data.info()

missing_values = data.isnull().sum()

missing_values
```

```python
data['date'] = pd.to_datetime(data['date'], errors='coerce')
data['laps'] = data['laps'].fillna(0).astype(int)

duplicates = data.duplicated().sum()

duplicates
```

```python
time_unique_values = data['time'].unique()

time_unique_values
```

```python
unique_winners = data['winner'].unique()
unique_cars = data['car'].unique()

unique_winners, unique_cars
```

```python
data['winner'] = data['winner'].str.strip()

data.head()
```

```python
def standardize_team_names(car_name):
    if 'Williams' in car_name:
        return 'Williams'
    elif 'Red Bull' in car_name or 'RBR' in car_name:
        return 'Red Bull Racing'
    elif 'McLaren' in car_name:
        return 'McLaren'
    elif 'Mercedes-Benz' in car_name:
        return 'Mercedes'
    else:
        return car_name

data['car'] = data['car'].apply(standardize_team_names)

standardized_cars = data['car'].unique()
standardized_cars
```

```python
team_wins = data['car'].value_counts()

driver_wins = data['winner'].value_counts()

top_teams = team_wins.head(10)
top_drivers = driver_wins.head(10)

top_teams, top_drivers
```

```python
data['year'] = data['date'].dt.year

team_wins_over_time = data.groupby(['year', 'car']).size().unstack(fill_value=0)
driver_wins_over_time = data.groupby(['year', 'winner']).size().unstack(fill_value=0)

cumulative_team_wins = team_wins_over_time[top_teams_list].cumsum()
cumulative_driver_wins = driver_wins_over_time[top_drivers_list].cumsum()
```

```python
top_teams_list = top_teams.index
top_drivers_list = top_drivers.index

def plot_cumulative_wins(data, title, xlabel, ylabel, colors, marker_interval=5):
    fig, ax = plt.subplots(figsize=(12, 8), facecolor='white')
    for i, entity in enumerate(data.columns):
        ax.plot(data.index, data[entity], label=entity, color=colors[i % len(colors)], linewidth=2, 
                marker='o', markevery=marker_interval)
    
    ax.set_title(title, fontsize=16, weight='bold')
    ax.set_xlabel(xlabel, fontsize=14)
    ax.set_ylabel(ylabel, fontsize=14)
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend(title='Entity', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
    plt.tight_layout()
    plt.show()

palette = sns.color_palette("tab10")
    
plot_cumulative_wins(cumulative_team_wins, "Cumulative Wins Over Time for Top Teams", "Year", "Cumulative Wins", palette, marker_interval=5)

plot_cumulative_wins(cumulative_driver_wins, "Cumulative Wins Over Time for Top Drivers", "Year", "Cumulative Wins", palette, marker_interval=5)
```

```python
def calculate_dominance_periods(wins_over_time, entity_type='Team'):
    dominance_periods = {}
    for entity in wins_over_time.columns:
        entity_wins = wins_over_time[entity]
        max_streak = 0
        current_streak = 0
        start_year = None
        end_year = None
        best_start_year = None
        best_end_year = None

        for year, wins in entity_wins.items():
            if wins > 0:
                current_streak += wins
                if current_streak == wins:
                    start_year = year
                end_year = year
                if current_streak > max_streak:
                    max_streak = current_streak
                    best_start_year = start_year
                    best_end_year = end_year
            else:
                current_streak = 0

        if max_streak > 0:
            dominance_periods[entity] = (max_streak, best_start_year, best_end_year)
    
    return dominance_periods

team_dominance_periods = calculate_dominance_periods(team_wins_over_time)

driver_dominance_periods = calculate_dominance_periods(driver_wins_over_time, entity_type='Driver')

top_team_dominance = {team: team_dominance_periods[team] for team in top_teams_list if team in team_dominance_periods}
top_driver_dominance = {driver: driver_dominance_periods[driver] for driver in top_drivers_list if driver in driver_dominance_periods}

top_team_dominance, top_driver_dominance
```

```python
def plot_dominance_periods(dominance_periods, title, xlabel, ylabel, colors):
    fig, ax = plt.subplots(figsize=(14, 8), facecolor='white')
    
    all_years = []
    for entity, (_, start_year, end_year) in dominance_periods.items():
        all_years.extend([start_year, end_year])
    min_year = min(all_years) - 1
    max_year = max(all_years) + 1
    
    for i, (entity, (wins, start_year, end_year)) in enumerate(dominance_periods.items()):
        bar_length = end_year - start_year + 1
        ax.barh(entity, bar_length, left=start_year, height=0.8, align='center', color=colors[i % len(colors)])
        ax.text(start_year + bar_length / 2, entity, f"{wins} wins", 
                va='center', ha='center', color='black', fontsize=12)
    
    ax.set_title(title, fontsize=16, weight='bold')
    ax.set_xlabel(xlabel, fontsize=14)
    ax.set_ylabel(ylabel, fontsize=14)
    ax.set_xlim(min_year, max_year)
    plt.tight_layout()
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

plot_dominance_periods(top_team_dominance, "Dominance Periods for Top Teams", "Years", "Teams", palette)

plot_dominance_periods(top_driver_dominance, "Dominance Periods for Top Drivers", "Years", "Drivers", palette)
```

```python
# Handling cases where time values might be NaN or improperly formatted
def safe_time_to_seconds(time_str):
    try:
        if pd.isna(time_str):
            return None
        h, m, s = 0, 0, 0.0
        parts = time_str.split(':')
        if len(parts) == 3:
            h, m, s = parts
        elif len(parts) == 2:
            m, s = parts
        return int(h) * 3600 + int(m) * 60 + float(s)
    except Exception as e:
        return None

# Apply the corrected function to the 'time' column and calculate average lap time
data['total_seconds'] = data['time'].apply(safe_time_to_seconds)
data = data.dropna(subset=['total_seconds', 'laps'])  # Drop rows with NaN in crucial columns
data['average_lap_time'] = data['total_seconds'] / data['laps']

# Group by year and calculate the running average of lap times
running_avg_lap_time = data.groupby('year')['average_lap_time'].mean().expanding().mean()

# Plot the running average of lap times over the years
plt.figure(figsize=(12, 8), facecolor='white')
plt.plot(running_avg_lap_time.index, running_avg_lap_time, label='Running Average Lap Time', color='blue')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Lap Time (seconds)', fontsize=14)
plt.title('Running Average of Lap Times by Race Over the Years', fontsize=16, weight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.animation as animation

# Prepare the data for animation
driver_wins_over_time = data.groupby(['year', 'winner']).size().unstack(fill_value=0).cumsum(axis=0)

# Initialize the figure and axis
fig, ax = plt.subplots(figsize=(12, 8), facecolor='white')

# Function to update the bar chart for each frame
def update(num):
    ax.clear()
    current_year = driver_wins_over_time.index[num]
    current_data = driver_wins_over_time.iloc[num]
    sorted_data = current_data.sort_values(ascending=False).head(10)
    ax.barh(sorted_data.index, sorted_data.values, color=palette[:10])
    ax.set_title(f'Cumulative Driver Wins - {current_year}', fontsize=16)
    ax.set_xlabel('Wins', fontsize=14)
    ax.set_xlim(0, sorted_data.max() + 5)
    ax.set_ylabel('Driver', fontsize=14)
    ax.grid(True, linestyle='--', alpha=0.7)

# Create the animation
ani = animation.FuncAnimation(fig, update, frames=len(driver_wins_over_time), repeat=False, interval=10)

# Save the animation to a file
ani.save('f1-data/driver_wins_race.mp4', writer='ffmpeg', fps=50)

# Display the animation
plt.show()
```
